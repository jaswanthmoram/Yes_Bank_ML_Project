## 3. ***Data Wrangling***

### Data Wrangling Code

````python

month_map = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
             'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

# Split the 'Date' column into 'month' and 'year'
df[['month', 'year']] = df['Date'].str.split('-', expand=True)

# Apply the month mapping
df['month'] = df['month'].map(month_map)

# Ensure 'year' is properly formatted to 4 digits
df['year'] = df['year'].apply(lambda x: '19' + x if int(x) > 50 else '20' + x)

# Convert 'year' column to integer type
df['year'] = pd.to_numeric(df['year'])

# Convert 'month' and 'year' into a proper datetime format
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

# Drop the original 'Date' column as it's no longer needed
df.drop('Date', axis=1, inplace=True)

# Display the first few rows
print(df.head())
````

##### 1. Why did you pick the specific chart?

The chosen charts (boxplot, KDE, and violin plot) help visualize the distribution, spread, and presence of outliers in the data. The boxplot highlights outliers, the KDE plot shows the density, and the violin plot combines both.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dummy DataFrame for visualization (excluding 'month')
df_viz = df.drop(columns=['month'], errors='ignore')

# Get numeric columns for visualization
numeric_cols = df_viz.select_dtypes(include=['number']).columns
num_cols = len(numeric_cols)

# Create subplots (3 plots per column: Boxplot, KDE, Violin)
fig, axes = plt.subplots(num_cols, 3, figsize=(15, 5 * num_cols))

# Flatten axes array for easy indexing
axes = axes.flatten()

# Loop through each numeric column and plot
for i, col in enumerate(numeric_cols):
    sns.boxplot(y=df_viz[col], ax=axes[i * 3])
    axes[i * 3].set_title(f'Boxplot of {col}')

    sns.kdeplot(df_viz[col], ax=axes[i * 3 + 1], fill=True)
    axes[i * 3 + 1].set_title(f'KDE Plot of {col}')

    sns.violinplot(y=df_viz[col], ax=axes[i * 3 + 2])
    axes[i * 3 + 2].set_title(f'Violin Plot of {col}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

1. Open and Close are positively correlated.
2. Low and High are positively correlated.

### Hypothetical Statement - 1 :Stock Prices Have Not Changed Significantly Over the Years

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis**: There is no significant difference in the mean opening price across different years.

**Alternate Hypothesis**: There is a significant difference in the mean opening price across different years.

#### 2. Perform an appropriate statistical test.

````python
import scipy.stats as stats

# Grouping opening prices by year
years = df['year'].unique()
groups = [df[df['year'] == y]['Open'] for y in years]

# Perform one-way ANOVA test
f_stat, p_value = stats.f_oneway(*groups)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Pearson's correlation test.

##### Why did you choose the specific statistical test?

ANOVA is used to compare means across multiple groups (in this case, different years).
If the p-value < 0.05, we reject the null hypothesis, meaning stock prices did change significantly over the years.

### Hypothetical Statement - 2 : Opening and Closing Prices Are Not Correlated

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis**: There is no correlation between the Opening Price and Closing Price of the stock.

````python
# Define CyclicalEncoder as it isn't present in sklearn.preprocessing
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline

class CyclicalEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, period=12):
        self.period = period  

    def fit(self, X, y=None):
        return self  

    def transform(self, X):
        X = X.to_numpy().astype(float).reshape(-1,1)
        X_sin = np.sin(2 * np.pi * X / self.period)
        X_cos = np.cos(2 * np.pi * X / self.period)
        return np.hstack((X_sin, X_cos))
    
pipeline = Pipeline([
    ('cyclical_encoder', CyclicalEncoder(period=12))
])
````

### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy.stats.mstats import winsorize
from scipy.stats import zscore

# Load dataset (Assuming df is already loaded)
df_cleaned = df.copy()  # Make a copy to avoid modifying the original DataFrame

# 1ï¸âƒ£ Handling Missing Values (Using Forward Fill, Backward Fill, and Interpolation)
df_cleaned.ffill(inplace=True)  # Forward Fill
df_cleaned.bfill(inplace=True)  # Backward Fill
df_cleaned.interpolate(method='linear', inplace=True)  # Linear Interpolation

# 2ï¸âƒ£ Handling Outliers (IQR Method & Winsorization)
Q1 = df_cleaned[['Open', 'High', 'Low', 'Close']].quantile(0.25)
Q3 = df_cleaned[['Open', 'High', 'Low', 'Close']].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df_cleaned[~((df_cleaned[['Open', 'High', 'Low', 'Close']] < (Q1 - 1.5 * IQR)) | 
                           (df_cleaned[['Open', 'High', 'Low', 'Close']] > (Q3 + 1.5 * IQR))).any(axis=1)]

df_cleaned.loc[:, 'Open'] = winsorize(df_cleaned['Open'], limits=[0.05, 0.05])  # Winsorization

# 3ï¸âƒ£ Feature Engineering
df_cleaned.loc[:, 'MA_5'] = df_cleaned['Close'].rolling(window=5).mean()  # 5-day moving average
df_cleaned.loc[:, 'MA_20'] = df_cleaned['Close'].rolling(window=20).mean()  # 20-day moving average
df_cleaned.loc[:, 'Price_Change'] = df_cleaned['Close'].pct_change()  # Percentage change
df_cleaned.loc[:, 'EMA_10'] = df_cleaned['Close'].ewm(span=10, adjust=False).mean()  # Exponential moving average
df_cleaned.loc[:, 'Log_Close'] = np.log(df_cleaned['Close'] + 1)  # Log Transformation

# 4ï¸âƒ£ Feature Scaling (Min-Max Scaling)
scaler = MinMaxScaler()
df_cleaned.loc[:, ['Open', 'High', 'Low', 'Close']] = scaler.fit_transform(df_cleaned[['Open', 'High', 'Low', 'Close']])

# 5ï¸âƒ£ Time-Based Features
df_cleaned.loc[:, 'Month'] = df_cleaned['month']
df_cleaned.loc[:, 'Year'] = df_cleaned['year']

# 6ï¸âƒ£ Lag Features
df_cleaned.loc[:, 'Close_Lag_1'] = df_cleaned['Close'].shift(1)  # Previous day's closing price
df_cleaned.loc[:, 'Close_Lag_5'] = df_cleaned['Close'].shift(5)  # Closing price from 5 days ago

# Display final processed dataset
print(df_cleaned.head())  # Check first few rows
````

### **Summary of Feature Manipulation**  

1. **Moving Averages (`MA_5`, `MA_20`)** â†’ Identifies **short-term and long-term trends** in stock prices.  
2. **Percentage Change (`Price_Change`)** â†’ Measures **stock price volatility** over time.  
3. **Exponential Moving Average (`EMA_10`)** â†’ Gives **more weight to recent prices**, making it more responsive.  
4. **Log Transformation (`Log_Close`)** â†’ **Normalizes skewed stock price data** for better analysis.  
5. **Min-Max Scaling** â†’ Rescales `Open`, `High`, `Low`, and `Close` to **standardize feature values**.  
6. **Time-Based Features (`Month`, `Year`)** â†’ Helps in analyzing **seasonal and yearly stock trends**.  
7. **Lag Features (`Close_Lag_1`, `Close_Lag_5`)** â†’ Uses past prices to **predict future stock movements**.  

These feature engineering techniques improve **data quality, trend analysis, and predictive accuracy**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Ensure stock_close has no missing values
stock_close = stock_close.dropna().reset_index(drop=True)

# Define Features (X) and Target (Y)
X = stock_close.drop(columns=['Close'])  # Features
Y = stock_close['Close']  # Target Variable

# Split dataset into Training (80%) and Testing (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize and Train Linear Regression Model
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)  # Model training

# Predict on Test Data
Y_pred = lin_model.predict(X_test)

# Evaluate Model Performance
r2 = r2_score(Y_test, Y_pred)  # R-squared score
mae = mean_absolute_error(Y_test, Y_pred)  # Mean Absolute Error

print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")

# Feature Importance Calculation (Fix)
feature_importance = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': lin_model.coef_})
feature_importance = feature_importance.reindex(feature_importance['Coefficient'].abs().sort_values(ascending=False).index)

# Print Feature Importance
print("\nFeature Importance Based on Linear Regression Coefficients:")
print(feature_importance)

# Plot Actual vs Predicted Stock Prices
plt.figure(figsize=(8, 5))
plt.scatter(Y_test, Y_pred, color='blue', alpha=0.6, label="Predicted vs Actual")
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--', label="Perfect Fit")
plt.xlabel("Actual Closing Price")
plt.ylabel("Predicted Closing Price")
plt.title("Linear Regression - Actual vs Predicted Stock Prices")
plt.legend()
plt.show()
````
